```python
# 1. Install the converter
# !pip install tensorflowjs

import tensorflow as tf
import tensorflowjs as tfjs
import numpy as np

# 2. Load Data (MNIST)
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize (0-255 -> 0-1) so it matches your web grid input
x_train, x_test = x_train / 255.0, x_test / 255.0

# Reshape to (28, 28, 1) to explicitly add the grayscale channel
x_train = x_train[..., np.newaxis]
x_test = x_test[..., np.newaxis]

# 3. Define a simple CNN Model
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(28, 28, 1)),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax') # Output: 10 probabilities
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 4. Train (5 epochs is usually enough for ~98% accuracy)
model.fit(x_train, y_train, epochs=5)

# 5. Export to TensorFlow.js format
tfjs.converters.save_keras_model(model, 'my_tfjs_model')

# 6. Download the files
from google.colab import files
import os

# Zip it up to download easily
!zip -r model.zip my_tfjs_model
files.download('model.zip')
```
