```python
import numpy as np
import pandas as pd
import statistics as stats
```

```python
arr = [12,45,67,23,89,55,34]
np_arr = np.array(arr)
print("Numpy Array:", np_arr)
```

```python
print("Mean:", np.mean(np_arr))
print("Median:", np.median(np_arr))
print("SD:", np.std(np_arr))
```

```python
student_data = {
    'Name' : ['Yash', 'Krunal', 'Manish', 'Raj', 'Parshwa'],
    'Marks' : [78,92,66,85,74],
    'Age' : [20, 21, 19, 22, 20]
}

df = pd.DataFrame(student_data)
print("Pandas DataFrame:\n", df)
```

```python
print("Marks Column:\n", df['Marks'])
```

```python
print("Pandas Describe Output:", df.describe())
```

```python
marks = df['Marks']

mean_marks = stats.mean(marks)
median_marks = stats.median(marks)
mode_marks = stats.mode(marks)
variance_marks = stats.variance(marks)

print(f"""
    Statistics using 'statistics' library:
    Mean     : {mean_marks}
    Median   : {median_marks}
    Mode     : {mode_marks}
    Variance : {variance_marks}
""")
```
