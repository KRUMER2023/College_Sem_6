```python
import pandas as pd
```

```python
# its names 'message' and 'label' as column name
msg = pd.read_csv('document.csv', names = ['message', 'label'])
msg.head()
```

```python
# printing dimension
print('The dimensions of the dataset', msg.shape)
```

--- 
### Create a new column 'labelnum' in msg dataframe by mapping:
- 'pos' to 1
- 'neg' to 0

```python

msg['labelnum'] = msg['label'].map({'pos' : 1, 'neg' : 0})
print(msg.head())
```

---
### Splitting the dataset into train-test data
default uses 75% train and 25% test 

```python
X = msg['message']
y = msg['labelnum']

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y)
```

```python
print("Training Data :", ytrain.shape)
print("Test Data :", ytest.shape)
print(xtrain)
print(xtest)
print(ytrain)
print(ytest)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cnt_vect = CountVectorizer()

xtrain_dtm = cnt_vect.fit_transform(xtrain)
print(xtrain_dtm)

xtest_dtm = cnt_vect.transform(xtest)
print(xtest_dtm)

print('Tokens in doc :')

print(cnt_vect.get_feature_names_out())
```
