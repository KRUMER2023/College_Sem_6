```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area --->')
plt.ylabel('Prices --->')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
# Predict price of a home with area = 3300 sqr ft
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Y = m * X + b (m is coefficient and b is intercept)
```

```python
3300*135.78767123 + 180616.43835616432 
```

```python
# The house price is given above for 3300 Square Feet Rs 628715.7534151643
```

```python
# Now Predict the price of a home with area = 5000 sqr ft
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
price = reg.coef_ * new_df + reg.intercept_ #This calculate prediction price using regression formula #Creates a smooth list of predicted prices to draw a straight regression line.
plt.plot(new_df, price, color='red', label='Linear Regression Line') #Exact Value of new_df and Predicted Price Values used to draw the smooth line
plt.legend()
plt.show()
```
